# Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("seaborn")
```

# Importing the dataset

```python
dataset = pd.read_csv("GlobalTerrorism.csv");
dataset.head()
```

```python
dataset.shape
```

# Data Preprocessing

```python
#Deleting unnecessary columns
df=dataset[["eventid","iyear","country_txt","region_txt","provstate","city","attacktype1_txt","targtype1_txt","gname","weaptype1_txt","nkill","nwound"]]
df.head()
```

```python
#Renaming columns
df=df.rename(columns={"eventid":"Event_id","iyear":"Year","country_txt":"Country","region_txt":"Region","provstate":"State","city":"City","attacktype1_txt":"Attack_type","targtype1_txt":"target_type","gname":"Group_name","weaptype1_txt":"Weapon_type","nkill":"Killed","nwound":"Wounded"})
df.head()
```

```python
#checking missing values
df.isnull().sum()
```

```python
#Handling missing values
df[["State","City"]]=df[["State","City"]].fillna("Unknown")
df[["Killed","Wounded"]]=df[["Killed","Wounded"]].fillna(0)
df.isnull().sum()
```

```python
#Creating a new column using wounded and killed columns
df["Casualties"]=df["Killed"]+df["Wounded"]
df.head()
```

```python
#Checking duplicate values
df.duplicated().sum()
```

# Exploratory Data Analysis

## Most Affected Countries

```python
#Number of unique countries
df["Country"].nunique()
```

```python
#Computing the terrorist attacks by countries and taking the top 15 observation
attacks_by_country=df.groupby("Country")["Event_id"].count().nlargest(15)
#Computing casualties in those top countries
Casualties_by_country=(df.groupby("Country")["Casualties"].sum())[attacks_by_country.index]

#visualizing number of attacks and casualties by country using bar chart
plt.figure(figsize=(15,7))
plt.barh(Casualties_by_country.index,Casualties_by_country,height=0.6,color="red")
plt.barh(attacks_by_country.index,attacks_by_country,height=0.6)
plt.title("Top 15 Most Affected Countries",fontsize=20)
plt.ylabel("Countries",fontsize=18)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.legend(["Number of Casualties","Number of Attacks"],fontsize=15)
plt.show()
```

#### Note: Iraq, Pakistan and Afganistan are affected by most of the attacks as well as casualties among the countries.

## Most Affected Regions

```python
#Number of unique regions
df["Region"].nunique()
```

```python
#Computing the terrorist attacks by regions and taking the top 10 observation
attacks_by_region=df.groupby("Region")["Event_id"].count().nlargest(10)
#Computing casualties in those top regions
Casualties_by_region=(df.groupby("Region")["Casualties"].sum())[attacks_by_region.index]

#visualizing attacks and casualties by Regions using bar chart

plt.figure(figsize=(15,7))
plt.barh(Casualties_by_region.index,Casualties_by_region,height=0.6,color="red")
plt.barh(attacks_by_region.index,attacks_by_region,height=0.6)
plt.title("Top 10 Most Affected Regions",fontsize=20)
plt.ylabel("Regions",fontsize=18)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.legend(["Number of Casualties","Number of Attacks"],fontsize=15)
plt.show()
```

#### Note: Among the regions, Middle East and North Africa is affected by most of the attacks as well as casualties.

## Most Affected States

```python
#Number of unique states
df["State"].nunique()
```

```python
#Computing the terrorist attacks by States and taking the top 20 observation
attacks_by_states=df.groupby("State")["Event_id"].count().nlargest(20)
#Computing casualties in those top states
Casualties_by_states=(df.groupby("State")["Casualties"].sum())[attacks_by_states.index]

#visualizing attacks and casualties by States using bar chart

plt.figure(figsize=(15,8))
plt.barh(Casualties_by_states.index,Casualties_by_states,height=0.6,color="red")
plt.barh(attacks_by_states.index,attacks_by_states,height=0.6)
plt.title("Top 20 Most Affected States",fontsize=20)
plt.ylabel("States",fontsize=18)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.legend(["Number of Casualties","Number of Attacks"],fontsize=15)
plt.show()
```

#### Note: Among the States, Baghdad is affected by most of the attacks as well as casualties.

## Year With Most Frequent Attacks

```python
#Number of unique years 
df["Year"].nunique()
```

```python
#Computing the terrorist attacks by Years and taking the top 10 observation
attacks_by_years=df.groupby("Year")["Event_id"].count()
#Computing casualties in those top years
Casualties_by_years=(df.groupby("Year")["Casualties"].sum())[attacks_by_years.index]

#visualizing attacks and casualties by years using bar chart

plt.figure(figsize=(20,8))
plt.bar(Casualties_by_years.index,Casualties_by_years,width=0.8,color="red")
plt.bar(attacks_by_years.index,attacks_by_years,width=0.8)
plt.title("Terrorist Attacks and Casualties by Year",fontsize=20)
plt.xlabel("Years",fontsize=18)
plt.xticks(attacks_by_years.index,rotation=90,fontsize=13)
plt.yticks(fontsize=13)
plt.legend(["Number of Casualties","Number of Attacks"],fontsize=15)
plt.show()
```

#### Note: Year 2014 has the maximum number of attacks as well as maximum number of casualties.

## Most Frequent Attack Type

```python
attack_type=df.groupby("Attack_type")["Event_id"].count().nlargest(10)

plt.figure(figsize=(15,8))

plt.bar(attack_type.index,attack_type,width=0.8)
plt.title("Most Frequent Types of Attack",fontsize=20)
plt.xlabel("Attack Types",fontsize=18)
plt.ylabel("Number of Attacks",fontsize=18)
plt.xticks(rotation=90,fontsize=13)
plt.yticks(fontsize=13)
plt.show()
```

#### Note: Bombing/Explotion is the most common type among the attack  types.

## Most Common Targets

```python
target_type=df.groupby("target_type")["Event_id"].count().nlargest(10)

plt.figure(figsize=(15,8))
plt.bar(target_type.index,target_type,width=0.8)
plt.title("Most Common Targets Types",fontsize=20)
plt.xlabel("Target Types",fontsize=18)
plt.ylabel("Number of Attacks",fontsize=18)
plt.xticks(rotation=90,fontsize=13)
plt.yticks(fontsize=13)
plt.show()
```

#### Note: Private Citizens and Properties are commonly targeted by terrorist groups.

## Most Common Weapon Type

```python
weapon_type=df.groupby("Weapon_type")["Event_id"].count().nlargest(10)


plt.figure(figsize=(15,8))
plt.bar(weapon_type.index,weapon_type,width=0.8)
plt.title("Most Common Weapon Types",fontsize=20)
plt.xlabel("Weapon Types",fontsize=18)
plt.ylabel("Number of Attacks",fontsize=18)
plt.xticks(rotation=90,fontsize=13)
plt.yticks(fontsize=13)
plt.show()
```

#### Note: Explosives are the most common types of weapons used for terrorist activities.

## Most Active Terrorist Groups

```python
terrorist_groups=df.groupby("Group_name")["Event_id"].count().nlargest(10)

plt.figure(figsize=(15,8))
plt.bar(terrorist_groups.index,terrorist_groups,width=0.8)
plt.title("Most Active Terrorist Groups",fontsize=20)
plt.xlabel("Terrorist Groups",fontsize=18)
plt.ylabel("Number of Attacks",fontsize=18)
plt.xticks(rotation=90,fontsize=13)
plt.yticks(fontsize=13)
plt.show()
```

#### Note: The second most active group is Taliban.
